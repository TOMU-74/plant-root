```python
import cv2
from matplotlib import pyplot as plt
```

### 画像の読みこみと色の変換

```python
img = cv2.imread('../original_sample.png')
seg = cv2.imread('../seg_sample.png')

#読み込めているか&画像サイズ（height, width, color）の確認
img.shape
```

↓ cv2では色はBGRで扱うが，matplotlibではRBGで扱うため，BとRが反転してしまう

```python
plt.imshow(img)
plt.show()
```

↓ cv2.cvtColorで BとRを変換させてから表示すると元通りの色になる

```python
img_ = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_)
plt.show()
```

```python
#画像の書き込み
cv2.imwrite('../seg_sample.png', img)
```
